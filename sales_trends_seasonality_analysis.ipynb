```python
dataset = pd.read_csv("2_Dimensional_sales_history_updated.csv", index_col=0)
dataset['transaction_date'] = pd.to_datetime(dataset['transaction_date'], infer_datetime_format=True)
IndexedDataset = dataset.set_index(['transaction_date'])
```

## Sum up rows with the same index i.e. sum transaction_total per month
## Dynamic sum

```python
dataset['transaction_total'].groupby(dataset['transaction_date'].dt.to_period('M')).sum().plot(kind='bar')
```

## Sum up transaction_total by day 

```python
df = dataset['transaction_total'].groupby(dataset['transaction_date'].dt.to_period('D')).sum().to_frame(name = 'sales').reset_index()
```

```python
df.mean()
```

```python
df.head()
```

```python
df.to_csv(r'../SciPY/sales_per_day.csv')
```

```python
df.plot()
```
